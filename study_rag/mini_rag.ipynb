```python
from langchain_ollama import ChatOllama
from dotenv import load_dotenv

load_dotenv()
```

```python
llm  = ChatOllama(model="deepseek-v3.1:671b-cloud",
                  base_url="https://ollama.com",
                  temperature=0.0)
```

```python
doc1 = "OpenAI была основана в декабре 2015 года. Цель компании – разработка ИИ для блага человечества."
doc2 = "МГУ – старейший университет в России, основанный в 1755 году. Расположен в Москве."
doc3 = "Париж называют городом любви. Он также знаменит своими историческими памятниками."
```

```python
docs = [doc1, doc2, doc3]
user_question = "В каком году был основан МГУ?"

relevant_docs = [d for d in docs if "основан" in d.lower() and ("мгу" in d.lower() or "университет" in d.lower())]

context = "\n".join(relevant_docs)  # просто склеим, в реале лучше брать кусочки
prompt = f"Вопрос: {user_question}\nВот данные из источников:\n{context}\nОтвет:"
prompt
```

```python
response = llm.invoke(prompt)
response.content
```
